```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
_fp = "C:\development\phinix-nexus-v1\src\outputs\pipeline_pipeline_visualizer_output_2024-07-08_1720467284.csv"
df = pd.read_csv(_fp)
```

```python
cols_to_stack = ['REQXXXXX_impacted', 'FORXXXXX_used', 'fields']
df_final = df.copy()
df_final = df.drop(columns=cols_to_stack)
df_final.drop_duplicates(ignore_index=True, inplace=True)
for col in cols_to_stack:
    df_stack = df[["ELEXXXXX_id", col]]
    df_stack.dropna(inplace=True, ignore_index=True)
    df_stack[col] = df_stack.groupby(['ELEXXXXX_id'])[col].transform(lambda x: ','.join(x))
    df_stack = df_stack[['ELEXXXXX_id',col]].drop_duplicates(ignore_index=True)
    df_final = pd.merge(df_final, df_stack, how="left", on=["ELEXXXXX_id"])
```

```python
from stock.xml_scripts.visualizer_parameters import GenerateVisualizerParameters
from load.write_xml import WriteXml

_generator = GenerateVisualizerParameters(params=df_final)
```

```python
def removeNoneAttrib(elem):
    """Remove None attributes from XML tree"""
    elem.attrib = {
        k: elem.attrib[k]
        for k in elem.attrib
        if elem.attrib[k] is not None
    }
    for subelem in elem:
        removeNoneAttrib(subelem)
```

```python
_sp = r"C:\development\phinix-nexus-v1\src\outputs\xml_test.xml"
_writer = WriteXml(file_path=_sp)

_xml_tree = _generator.generate_xml()
removeNoneAttrib(_xml_tree.getroot())
_writer.write_file(output=_xml_tree)
```
